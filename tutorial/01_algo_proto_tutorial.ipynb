## 1. 类初始化

```python
# 导入 AlgoProto
from algodisco.base.algo import AlgoProto

# 方法1: 直接初始化
algo1 = AlgoProto(
    program="def sort_list(lst):\n    return sorted(lst)",
    language="python",
    score=0.95
)
print(f"algo_id: {algo1.algo_id}")
print(f"program: {algo1.program[:30]}...")
print(f"language: {algo1.language}")
print(f"score: {algo1.score}")
```

```python
# 方法2: 从程序字符串创建 (from_program_str)
algo2 = AlgoProto.from_program_str(
    "def quick_sort(arr):\n    return sorted(arr)"
)
print(f"自动推断 language: {algo2.language}")
print(f"默认 score: {algo2.score}")
```

```python
# 方法3: 从字典创建 (from_dict) - 最常用
data = {
    "program": "def bubble_sort(lst):\n    # 实现冒泡排序\n    n = len(lst)\n    for i in range(n):\n        for j in range(0, n-i-1):\n            if lst[j] > lst[j+1]:\n                lst[j], lst[j+1] = lst[j+1], lst[j]\n    return lst",
    "language": "python",
    "score": 0.85,
    "execution_time": 0.023,
    "prompt_tokens": 1200,
}

algo3 = AlgoProto.from_dict(data)
print(f"algo_id: {algo3.algo_id}")
print(f"score: {algo3.score}")
print(f"元数据: {algo3.metadata}")
```

## 2. 基本属性

```python
# AlgoProto 的5个核心属性
algo = AlgoProto(
    program="print('hello')",
    language="python",
    score=1.0
)

# algo_id: 唯一标识符（自动生成UUID）
print(f"algo_id: {algo.algo_id}")

# program: 算法代码
print(f"program: {algo.program}")

# language: 编程语言
print(f"language: {algo.language}")

# score: 评分（可以是任意类型）
print(f"score: {algo.score}")

# metadata: 字典，存储额外信息
print(f"metadata: {algo.metadata}")
```

## 3. 字典互转

```python
# to_dict: 转换为字典
algo = AlgoProto.from_dict({
    "program": "def add(a, b):\n    return a + b",
    "language": "python",
    "score": 0.9,
    "execution_time": 0.01,
    "prompt": "Generate an add function"
})

d = algo.to_dict()
print("转换为字典:")
for k, v in d.items():
    print(f"  {k}: {v}")
```

```python
# update: 从字典更新属性
algo = AlgoProto(
    program="def test(): pass",
    language="python"
)

algo.update({
    "score": 0.75,
    "execution_time": 0.05,
    "error_msg": None
})

print(f"score: {algo.score}")
print(f"metadata: {algo.metadata}")
```

## 4. 元数据操作

```python
# 使用 [] 语法操作 metadata
algo = AlgoProto(program="x = 1", language="python")

# 设置元数据
algo["execution_time"] = 0.023
algo["prompt_tokens"] = 1500
algo["response_tokens"] = 300
algo["model"] = "gpt-4"

# 获取元数据
print(f"execution_time: {algo['execution_time']}")
print(f"prompt_tokens: {algo['prompt_tokens']}")

# get 方法 - 带默认值的获取
print(f"不存在时返回默认值: {algo.get('temperature', 0.7)}")
print(f"存在时返回实际值: {algo.get('model', 'gpt-3.5')}")
```

```python
# pop 方法 - 取出并删除
algo = AlgoProto(program="x = 1", language="python")
algo["temp_data"] = "some_value"

value = algo.pop("temp_data")
print(f"pop出的值: {value}")
print(f"pop后metadata: {algo.metadata}")

# pop 带默认值
value = algo.pop("not_exists", "default_value")
print(f"不存在的key: {value}")
```

```python
# keep_metadata_keys - 只保留指定的keys
algo = AlgoProto.from_dict({
    "program": "code",
    "score": 0.9,
    "execution_time": 0.1,
    "prompt": "test prompt",
    "response": "test response",
    "error_msg": None
})

print("保留前:", algo.metadata)

algo.keep_metadata_keys(["execution_time", "score"])

print("保留后:", algo.metadata)
```

## 5. 深度拷贝

```python
import copy

# __deepcopy__ - 自动生成新的algo_id
algo1 = AlgoProto.from_dict({
    "program": "def test(): pass",
    "language": "python",
    "score": 0.8,
    "execution_time": 0.01
})

algo2 = copy.deepcopy(algo1)

print(f"原 algo_id: {algo1.algo_id}")
print(f"拷贝 algo_id: {algo2.algo_id}")
print(f"score 相同: {algo1.score == algo2.score}")
print(f"metadata 相同: {algo1.metadata == algo2.metadata}")
```

## 6. 实用技巧

```python
# 技巧1: 获取Markdown格式的代码块
algo = AlgoProto(
    program="def hello():\n    print('world')",
    language="python"
)

print(algo.get_markdown_code_block())
```

```python
# 技巧2: 验证属性访问
algo = AlgoProto(program="x = 1", language="python", score=0.5)

# 可以通过 get 方法访问所有属性（包括metadata）
print(f"get('program'): {algo.get('program')}")
print(f"get('score'): {algo.get('score')}")
print(f"get('不存在'): {algo.get('不存在', '默认值')}")
```

```python
# 技巧3: 批量创建 AlgoProto
codes = [
    "def sort1(arr): return sorted(arr)",
    "def sort2(arr): return sorted(arr)",
    "def sort3(arr): return sorted(arr)",
]

algos = [AlgoProto.from_dict({
    "program": code,
    "language": "python",
    "score": 0.5 + i * 0.1
}) for i, code in enumerate(codes)]

for a in algos:
    print(f"id={a.algo_id[:8]}... score={a.score}")
```

```python
# 技巧4: AlgoProto 列表按 score 排序
algos = [
    AlgoProto.from_dict({"program": "c", "score": 0.3}),
    AlgoProto.from_dict({"program": "a", "score": 0.9}),
    AlgoProto.from_dict({"program": "b", "score": 0.5}),
]

# 按 score 降序排序
sorted_algos = sorted(algos, key=lambda x: x.score, reverse=True)

print("排序后:")
for a in sorted_algos:
    print(f"  score={a.score}, program={a.program}")
```

## 总结

`AlgoProto` 核心要点:

| 方法 | 用途 |
|------|------|
| `from_program_str()` | 从代码字符串创建 |
| `from_dict()` | 从字典创建（推荐） |
| `to_dict()` | 转换为字典 |
| `update()` | 从字典更新 |
| `get_markdown_code_block()` | 获取Markdown格式 |
| `keep_metadata_keys()` | 保留指定keys |
| `copy.deepcopy()` | 深度拷贝（自动生成新ID）|
