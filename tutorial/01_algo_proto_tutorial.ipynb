# Complete Guide to AlgoProto

`AlgoProto` is the core data structure used to represent algorithms in AlgoDisco. It stores program code, scores, metadata, and the identifier used to track each candidate during search.

## 1. Initialization


```python
# Import AlgoProto
from algodisco.base.algo import AlgoProto

# Method 1: Direct initialization
algo1 = AlgoProto(
    program="def sort_list(lst):\n    return sorted(lst)",
    language="python",
    score=0.95
)
print(f"algo_id: {algo1.algo_id}")
print(f"program: {algo1.program[:30]}...")
print(f"language: {algo1.language}")
print(f"score: {algo1.score}")
```

```python
# Method 2: Create from a program string (from_program_str)
algo2 = AlgoProto.from_program_str(
    "def quick_sort(arr):\n    return sorted(arr)"
)
print(f"inferred language: {algo2.language}")
print(f"default score: {algo2.score}")
```

```python
# Method 3: Create from a dictionary (from_dict) - the most common pattern
data = {
    "program": "def bubble_sort(lst):\n    # Bubble sort implementation\n    n = len(lst)\n    for i in range(n):\n        for j in range(0, n-i-1):\n            if lst[j] > lst[j+1]:\n                lst[j], lst[j+1] = lst[j+1], lst[j]\n    return lst",
    "language": "python",
    "score": 0.85,
    "execution_time": 0.023,
    "prompt_tokens": 1200,
}

algo3 = AlgoProto.from_dict(data)
print(f"algo_id: {algo3.algo_id}")
print(f"score: {algo3.score}")
print(f"metadata: {algo3.metadata}")
```

## 2. Core Attributes


```python
# The 5 core AlgoProto attributes
algo = AlgoProto(
    program="print('hello')",
    language="python",
    score=1.0
)

# algo_id: unique identifier (UUID generated automatically)
print(f"algo_id: {algo.algo_id}")

# program: algorithm source code
print(f"program: {algo.program}")

# language: programming language
print(f"language: {algo.language}")

# score: algorithm score (can be any type)
print(f"score: {algo.score}")

# metadata: dictionary for additional information
print(f"metadata: {algo.metadata}")
```

## 3. Dictionary Conversion


```python
# to_dict: convert to a dictionary
algo = AlgoProto.from_dict({
    "program": "def add(a, b):\n    return a + b",
    "language": "python",
    "score": 0.9,
    "execution_time": 0.01,
    "prompt": "Generate an add function"
})

d = algo.to_dict()
print("Converted dictionary:")
for k, v in d.items():
    print(f"  {k}: {v}")
```

```python
# update: update attributes from a dictionary
algo = AlgoProto(
    program="def test(): pass",
    language="python"
)

algo.update({
    "score": 0.75,
    "execution_time": 0.05,
    "error_msg": None
})

print(f"score: {algo.score}")
print(f"metadata: {algo.metadata}")
```

## 4. Metadata Operations


```python
# Use [] syntax to work with metadata
algo = AlgoProto(program="x = 1", language="python")

# Set metadata
algo["execution_time"] = 0.023
algo["prompt_tokens"] = 1500
algo["response_tokens"] = 300
algo["model"] = "gpt-4o-mini"

# Read metadata
print(f"execution_time: {algo['execution_time']}")
print(f"prompt_tokens: {algo['prompt_tokens']}")

# get: read with a default value
print(f"missing key returns default: {algo.get('temperature', 0.7)}")
print(f"existing key returns actual value: {algo.get('model', 'fallback-model')}")
```

```python
# pop: fetch and delete a metadata value
algo = AlgoProto(program="x = 1", language="python")
algo["temp_data"] = "some_value"

value = algo.pop("temp_data")
print(f"popped value: {value}")
print(f"metadata after pop: {algo.metadata}")

# pop with a default value
value = algo.pop("not_exists", "default_value")
print(f"missing key: {value}")
```

```python
# keep_metadata_keys: keep only selected keys
algo = AlgoProto.from_dict({
    "program": "code",
    "score": 0.9,
    "execution_time": 0.1,
    "prompt": "test prompt",
    "response": "test response",
    "error_msg": None
})

print("before filtering:", algo.metadata)

algo.keep_metadata_keys(["execution_time", "score"])

print("after filtering:", algo.metadata)
```

## 5. Deep Copy


```python
import copy

# __deepcopy__ automatically creates a new algo_id
algo1 = AlgoProto.from_dict({
    "program": "def test(): pass",
    "language": "python",
    "score": 0.8,
    "execution_time": 0.01
})

algo2 = copy.deepcopy(algo1)

print(f"original algo_id: {algo1.algo_id}")
print(f"copied algo_id: {algo2.algo_id}")
print(f"same score: {algo1.score == algo2.score}")
print(f"same metadata: {algo1.metadata == algo2.metadata}")
```

## 6. Practical Tips


```python
# Tip 1: Get a Markdown-formatted code block
algo = AlgoProto(
    program="def hello():\n    print('world')",
    language="python"
)

print(algo.get_markdown_code_block())
```

```python
# Tip 2: Verify attribute access
algo = AlgoProto(program="x = 1", language="python", score=0.5)

# get can access built-in attributes as well as metadata
print(f"get('program'): {algo.get('program')}")
print(f"get('score'): {algo.get('score')}")
print(f"get('missing'): {algo.get('missing', 'default value')}")
```

```python
# Tip 3: Create many AlgoProto objects in a batch
codes = [
    "def sort1(arr): return sorted(arr)",
    "def sort2(arr): return sorted(arr)",
    "def sort3(arr): return sorted(arr)",
]

algos = [AlgoProto.from_dict({
    "program": code,
    "language": "python",
    "score": 0.5 + i * 0.1
}) for i, code in enumerate(codes)]

for a in algos:
    print(f"id={a.algo_id[:8]}... score={a.score}")
```

```python
# Tip 4: Sort a list of AlgoProto objects by score
algos = [
    AlgoProto.from_dict({"program": "c", "score": 0.3}),
    AlgoProto.from_dict({"program": "a", "score": 0.9}),
    AlgoProto.from_dict({"program": "b", "score": 0.5}),
]

# Sort by score in descending order
sorted_algos = sorted(algos, key=lambda x: x.score, reverse=True)

print("sorted results:")
for a in sorted_algos:
    print(f"  score={a.score}, program={a.program}")
```

## Summary

Key `AlgoProto` operations:

| Method | Purpose |
|------|------|
| `from_program_str()` | Create from a program string |
| `from_dict()` | Create from a dictionary (recommended) |
| `to_dict()` | Convert to a dictionary |
| `update()` | Update from a dictionary |
| `get_markdown_code_block()` | Render as a Markdown code block |
| `keep_metadata_keys()` | Keep only selected metadata keys |
| `copy.deepcopy()` | Deep copy and generate a new ID |
